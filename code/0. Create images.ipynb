```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
import os
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
directory = Path("/home/namit/codes/Entropy-Isomap/selected_data/square/") 
target_directory = Path("/home/namit/codes/Entropy-Isomap/outputs/images_square/")
selected_trajectories = ['BR0.52-CHI2.4-R1', 'BR0.52-CHI2.4-R3', 'BR0.56-CHI2.4-R1', 'BR0.56-CHI2.4-R4', 'BR0.54-CHI2.4-R5']
```

```python
for trajectory in tqdm(filter(lambda x: x.name[:2]=='BR', directory.iterdir())):
    target_trajectory_directory = target_directory/trajectory.name
    target_trajectory_directory.mkdir(parents=True, exist_ok=True)
    if trajectory.name in selected_trajectories:
        print(trajectory.name)
        for index,point in enumerate(sorted(filter(lambda x: x.name[:8]=='dataCH.0',trajectory.iterdir()))):
            print(point.name)
            img_data = np.genfromtxt(point, skip_header=3, skip_footer=159201)
            img = np.reshape(img_data[:,2], (400,400))
            plt.imsave(target_trajectory_directory/(str(index)+'.png'),img,cmap='gray')
```
